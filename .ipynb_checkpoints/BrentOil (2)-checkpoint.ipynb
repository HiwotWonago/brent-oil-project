```python
prices = df['Price'].values
n_days = len(prices)
```

```python
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
import arviz as az

df['log_return'] = np.log(df['Price'] / df['Price'].shift(1))

# Data Preparation
log_returns = df['log_return'].dropna().values
n = len(log_returns)
x = np.arange(n)
# Step 2: model-development
with pm.Model() as model:
    tau = pm.DiscreteUniform("tau", lower=0, upper=n)
    mu_1 = pm.Normal("mu_1", mu=log_returns.mean(), sigma=1)
    mu_2 = pm.Normal("mu_2", mu=log_returns.mean(), sigma=1)
    sigma_1 = pm.HalfNormal("sigma_1", sigma=1)
    sigma_2 = pm.HalfNormal("sigma_2", sigma=1)
    mu = pm.math.switch(x < tau, mu_1, mu_2)
    sigma = pm.math.switch(x < tau, sigma_1, sigma_2)
    obs = pm.Normal("obs", mu=mu, sigma=sigma, observed=log_returns)
    # Step 3: Sampling
    trace = pm.sample(2000, tune=1000, target_accept=0.95, return_inferencedata=True, random_seed=42)
```

```python
# Step 4: Trace summary
az.summary(trace, var_names=["tau", "mu_1", "mu_2", "sigma_1", "sigma_2"])
```

```python
# Step 5: Trace plot
az.plot_trace(trace, var_names=["tau", "mu_1", "mu_2"])
plt.show()
```

```python
# Step 6: Posterior of change point
plt.figure(figsize=(10, 4))
plt.title("Posterior Distribution of Change Point (tau)")
az.plot_posterior(trace, var_names=["tau"])
plt.axvline(trace.posterior["tau"].mean(), color="red", linestyle="--", label="Mean tau")
plt.legend()
plt.show()
```

```python
# Get the most probable change point
tau_map = int(trace.posterior['tau'].mean().values)
# Plot log return series with the change point
plt.figure(figsize=(12, 4))
plt.plot(df.index[1:], log_returns, label="Log Returns", color="gray")
plt.axvline(df.index[tau_map + 1], color="red", linestyle="--", label="Change Point")
plt.title("Log Returns with Detected Change Point")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
plt.plot(df['Price'], label='Brent Oil Price', color='black')
plt.title("Brent Oil Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import os

# Create the 'data' directory if it doesn't exist
if not os.path.exists('data'):
    os.makedirs('data')

# Create a DataFrame for the change point
# We need the date and price at the detected change point (tau_map)
# Remember that tau_map is an index, and the log_returns were calculated from the second data point,
# so the index in the original df for the change point is tau_map + 1
change_point_date = df['Date'].iloc[tau_map + 1]
change_point_price = df['Price'].iloc[tau_map + 1]

change_points_df = pd.DataFrame({
    'Date': [change_point_date],
    'Price': [change_point_price]
})

# Save the change points DataFrame to a CSV file in the 'data' directory
change_points_df.to_csv('data/change_points.csv', index=False)

print("Created 'data/change_points.csv' with the detected change point.")
display(change_points_df)
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
from dash import Dash, dcc, html, Input, Output

# Load Brent oil data with log returns and change points
df = pd.read_csv('data/brent_oil_log_returns.csv', parse_dates=['Date'])
change_points = pd.read_csv('data/change_points.csv', parse_dates=['Date'])

# Initialize Dash app
app = Dash(__name__)
app.title = "Brent Oil Analysis Dashboard"

app.layout = html.Div([
    html.H1("📈 Brent Oil Price Dashboard", style={'textAlign': 'center'}),

    dcc.DatePickerRange(
        id='date-picker',
        start_date=df['Date'].min(),
        end_date=df['Date'].max(),
        display_format='YYYY-MM-DD'
    ),

    dcc.Graph(id='price-graph'),
    dcc.Graph(id='log-return-graph'),

    html.Div("Developed with Plotly Dash", style={'textAlign': 'center', 'padding': '20px'})
])

@app.callback(
    [Output('price-graph', 'figure'),
     Output('log-return-graph', 'figure')],
    [Input('date-picker', 'start_date'),
     Input('date-picker', 'end_date')]
)
def update_graphs(start_date, end_date):
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    filtered = df.loc[mask]
    cp_filtered = change_points.loc[(change_points['Date'] >= start_date) & (change_points['Date'] <= end_date)]

    # Raw Price Chart with Change Points
    price_fig = go.Figure()
    price_fig.add_trace(go.Scatter(x=filtered['Date'], y=filtered['Price'], mode='lines', name='Price'))
    price_fig.add_trace(go.Scatter(
        x=cp_filtered['Date'], y=cp_filtered['Price'],
        mode='markers', marker=dict(size=10, color='red'),
        name='Change Points'
    ))
    price_fig.update_layout(title='Brent Oil Price Over Time', xaxis_title='Date', yaxis_title='Price')

    # Log Returns Chart
    log_fig = go.Figure()
    log_fig.add_trace(go.Scatter(x=filtered['Date'], y=filtered['LogReturn'], mode='lines', name='Log Return'))
    log_fig.update_layout(title='Log Returns Over Time', xaxis_title='Date', yaxis_title='Log Return')

    return price_fig, log_fig

if __name__ == '__main__':
    app.run(debug=True)
```
